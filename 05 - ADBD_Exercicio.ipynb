```python
# Importar biblioteca:
import pyodbc   # Conexão com o BD
import ctypes # Caixa de mensagens

# Classe para conexão com o banco de dados:
class ConexaoBD ():
    def conectar (self):
        return pyodbc.connect (driver = '{SQL Server}', server = 'DESKTOP-ATE5M9D\SQLEXPRESS', database ='empresa', trusted_connection = 'yes')

# Instanciar um objeto da classe para conexão com o banco de dados:
bd = ConexaoBD ()

# Pegar uma conexão com o banco de dados:
conexao = bd.conectar ()
```

```python
# Classe Categoria:
class Categoria:
    def __init__ (self, conexao):
        self.conexao = conexao
    def Criar (self):
        ddl = '''if not exists (select * from sysobjects where name='Categoria' and xtype='U')
                    create table Categoria ( 
                                id int primary key identity (1, 1), -- identity (inicio, incremento): especifica o valor inicial e de incremento.
                                nome varchar (32),
                                descricao text
                                ); '''
        with self.conexao:
            self.conexao.execute (ddl)

#   EXECUTAR ROTINAS
    def Executar(self):
        resp = input ("O que deseja fazer? 1 - Inserir  \n  2 - Atualizar  \n  3 - Consultar  |  4 - Excluir |  0 - Para/Sair")
        print(resp)
        if (resp == '1'):
            # Inserir um registro no banco de dados:
            linha = self.Inserir ()
            # Apresentar na tela o registro inserido no banco de dados:
            ret = False
        elif(resp == '2'):
            self.Atualizar()
            # Apresentar na tela o registro inserido no banco de dados:
            ret = False
        elif(resp == '3'):
            linha = self.Consultar()
            # Apresentar na tela o registro inserido no banco de dados:
            ret = False
        elif(resp == '4'):
            self.Excluir()
            # Apresentar na tela o registro inserido no banco de dados:
            ret = False            
        elif(resp == '0'):
            ctypes.windll.user32.MessageBoxW(0, "Finalizando aplicação, até mais!","Menssagem", 0)
            ret = True 
        else:            
            ctypes.windll.user32.MessageBoxW(0, "Opção Invalida!","Alerta", 0)
            ret = False

        return ret
#   COLETA DE VALORES
    def Valores (self,varios = True):
            parametros = []
            while True:
                nome = input ("Informe o nome da categoria: ") 
                descricao = input ("Informe a descrição da categoria: ")
                parametros.append ((nome, descricao))
                print (parametros)

                if(varios == True):
                    resp = input ("Deseja informar outro registro? (s/n)")
                    if (resp == "n"):
                        break
                else:
                    break

            return parametros

#   1 - INSERT        
    def Inserir (self):
        dml = "insert into Categoria (nome, descricao) values (?, ?);"
        parametros = self.Valores ()
        with self.conexao:
            cursor = self.conexao.cursor ()
            cursor.executemany (dml, parametros)
            # self.conexao.commit ()
            sql = "select * from Categoria;"
            cursor.execute (sql)
            for nome in cursor:
               print (nome)
        
 #  2 - ATUALIZAR
    def Atualizar (self):
        id = input ("Informe o id da categoria que será atualizada: ")
        dml = "Update Categoria set nome = ?, descricao = ? Where id = "+ id +";"
        parametros = self.Valores ()
        with self.conexao:
            cursor = self.conexao.cursor ()
            cursor.executemany (dml, parametros)
            # self.conexao.commit ()
            sql = "select * from Categoria;"
            cursor.execute (sql)
            for nome in cursor:
               print (nome)
        
 #  3 - CONSULTAR
    def Consultar (self, Semfiltro = False):
        sql = "Select * from Categoria"
        with self.conexao:
            if (Semfiltro == False):
                resp = ctypes.windll.user32.MessageBoxW(0, "Deseja consultar um registro especifico?", "Atenção",  3)
                if(resp == 6):
                    id = input ("Informe o id da categoria que será consultada: ")
                    sql = sql + ' where id  = ' + id + ';'
                else:
                    sql = sql + ';'
            else:
                sql = sql + ';'    
            cursor = self.conexao.cursor ()
            cursor.execute (sql)
            for nome in cursor:
               print (nome)
 #  4 - EXCLUIR
    def Excluir (self):       
        with self.conexao:
            id = input ("Informe o código da categoria que será atualizada: ")
            sql = "Delete from Categoria where id  = " + id + ";"
            cursor = self.conexao.cursor ()
            cursor.execute (sql)

            sql = "select * from Categoria;"
            cursor.execute (sql)
            for nome in cursor:
               print (nome)
```
